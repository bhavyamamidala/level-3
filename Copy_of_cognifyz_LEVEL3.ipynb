**Task1: Predictive Modeling**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

Build a regression model to predict the
aggregate rating of a restaurant based on
available features.
Split the dataset into training and testing sets
and evaluate the model's performance using
appropriate metrics.

```python
df = pd.read_csv('/content/drive/MyDrive/Dataset.csv',encoding= 'unicode_escape')
```

```python
df['Has Table Booking'] = df['Has Table booking'].map({'Yes': 1, 'No': 0})
df['Has Online Delivery'] = df['Has Online delivery'].map({'Yes': 1, 'No': 0})
print(df)
```

```python
features = df[['Average Cost for two', 'Has Table Booking', 'Has Online Delivery', 'Votes']]
target = df['Aggregate rating']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

Experiment with different algorithms (e.g.,
linear regression, decision trees, random
forest) and compare their performance

```python

features = df[['Average Cost for two', 'Has Table Booking', 'Has Online Delivery', 'Votes']]
target = df['Aggregate rating']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
y_pred_linear = linear_model.predict(X_test)
mse_linear = mean_squared_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)
print("Linear Regression - Mean Squared Error:", mse_linear)
print("Linear Regression - R-squared:", r2_linear)
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)
print("Decision Tree Regression - Mean Squared Error:", mse_dt)
print("Decision Tree Regression - R-squared:", r2_dt)
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest Regression - Mean Squared Error:", mse_rf)
print("Random Forest Regression - R-squared:", r2_rf)
```

**Task2: Customer Preference Analysis**

Analyze the relationship between the type of
cuisine and the restaurant's rating.


```python
average_rating_by_cuisine = df.groupby('Cuisines')['Aggregate rating'].mean()
sorted_cuisines = average_rating_by_cuisine.sort_values(ascending=False)
print("Top Cuisines with the Highest Average Rating:")
print(sorted_cuisines.head(10))
```

Identify the most popular cuisines among
customers based on the number of votes.


```python
total_votes_by_cuisine = df.groupby('Cuisines')['Votes'].sum()
sorted_cuisines_by_votes = total_votes_by_cuisine.sort_values(ascending=False)
print("Top Cuisines with the Highest Total Votes:")
print(sorted_cuisines_by_votes.head(10))
```

Determine if there are any specific cuisines
that tend to receive higher ratings.

```python
average_rating_by_cuisine = df.groupby('Cuisines')['Aggregate rating'].mean()
sorted_cuisines_by_rating = average_rating_by_cuisine.sort_values(ascending=False)
print("Top Cuisines with the Highest Average Rating:")
print(sorted_cuisines_by_rating.head(10))
```

**Task3: Data Visualization**

**Create visualizations to represent the distribution
of ratings using different charts (histogram, bar
plot, etc.).**

```python
#Create a histogram to visualize the distribution of ratings
plt.figure(figsize=(8, 6))
plt.hist(df['Aggregate rating'], bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()
```

```python

# Create a bar plot to visualize the average rating for each price range
plt.figure(figsize=(10, 6))
sns.barplot(x='Price range', y='Aggregate rating', data=df, ci=None, palette='viridis')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.title('Average Rating by Price Range')
plt.show()
```

```python
# Create a box plot to visualize the distribution of ratings for each price range
plt.figure(figsize=(10, 6))
sns.boxplot(x='Price range', y='Aggregate rating', data=df, palette='Set2')
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')
plt.title('Distribution of Ratings by Price Range')
plt.show()
```

**Compare the average ratings of different cuisines
or cities using appropriate visualizations.**

```python
average_rating_by_cuisine = df.groupby('Cuisines')['Aggregate rating'].mean().reset_index()
sorted_cuisines_by_rating = average_rating_by_cuisine.sort_values(by='Aggregate rating', ascending=False)
top_n_cuisines = 10
top_cuisines = sorted_cuisines_by_rating.head(top_n_cuisines)
plt.figure(figsize=(12, 6))
sns.barplot(x='Cuisines', y='Aggregate rating', data=top_cuisines, palette='viridis')
plt.xlabel('Cuisines')
plt.ylabel('Average Rating')
plt.title(f'Top {top_n_cuisines} Cuisines with Highest Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
